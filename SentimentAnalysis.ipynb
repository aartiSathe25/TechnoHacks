```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words='english')
words = cv.fit_transform(train.text)
sum_words = words.sum(axis=0)

words_freq = [(words, sum_words[0,i]) for words, i in cv.vocabulary_.items()]
words_freq = sorted(words_freq, key=lambda x:x[1], reverse = True)
freq = pd.DataFrame(words_freq, columns=['word','freq'])
freq.head(30).plot(x='word',y='freq',kind='bar',figsize=(15,7),color='blue')
plt.title('Most Frquently Occuring Words - Top 30')
```
